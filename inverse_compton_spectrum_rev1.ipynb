```python
def integrate_diff_spectrum(x,y):
    I = np.zeros(len(y));
    
    y[np.isnan(y)] = 0;
    
    for i in range(0, len(y)):
        I[i] = np.trapz(y[i:len(y)], x=x[i:len(y)]);
    return I;
```

# Solar Blackbody Spectrum

Our starting point is the power spectral density of the Sun, assuming blackbody radiation:

$$S_\lambda = \frac{2\pi c^2 h}{\lambda^5} \frac{1}{e^{hc/\lambda kT} - 1}~~ [W/m^2/nm]$$

and we'll impose that the integration of $S_\lambda$ over wavelength must be equal to the total spectral irradiance at Earth:

$$I_{Sun,~Earth} = \int_\lambda S_\lambda d\lambda = 1360.8 ~W/m^2$$

We use the following relationship to scale the flux from the Sun to Jupiter:

$$I_{Sun,~Jupiter} = I_{Sun,~Earth} \cdot \left(\frac{1~AU}{5~AU}\right)^2$$

We are interested in the individual photons at each wavelength, so we divide by the energy per photon, $E = hc/\lambda$ [J/photon]:

$$N_\lambda = S_\lambda \cdot \frac{\lambda}{hc} ~~[photons/m^2/s/nm]$$

And integrate over infinitesmal wavelength bins $d\lambda$ to obtain the continuum number of photons at each wavelength:

$$N = \int_{\lambda'}^\infty N_\lambda \cdot~d\lambda = \int_{\lambda'}^\infty S_\lambda \cdot \frac{\lambda}{hc} \cdot~d\lambda ~~~ [photons/m^2-s]$$

To get the volumetric number density of photons at Jupiter, we assume that in 1 second the photons (which at 5 AU are a parallel beam from an infinitely far point source) transit a box the size of $\Delta x = c\Delta t = 3\times10^8~m$, where $\Delta t$ is unity. We'll then divide by the box side length and multiply by the duration of unity to obtain

$$N_{ph} = \Delta t\cdot\frac{1}{c\Delta t}N = \frac{1}{c}N~~~ [photons/m^3]$$

```python
wavelength = np.linspace(100, 5000, 1000) * 1e-9; # nm -> m
T_sun      = 5778; # K
I_sun      = 1360.8; # W/m^2

solar_spectra = 2*np.pi * const.c**2*const.h/wavelength**5 * \
                1/(np.exp(const.h*const.c/(wavelength*const.k*T_sun))-1)

solar_spectra /= np.trapz(y=solar_spectra, x=wavelength*1e9)
solar_spectra *= I_sun

J_solar_spectra = solar_spectra * (1/5)**2

plt.figure(figsize=(16,4)); plt.subplot(1,2,1); 
plt.plot(wavelength*1e9, solar_spectra)
plt.xlabel('$\lambda$ [nm]', fontsize=14);
plt.ylabel('Power Spectral Density\nS$_\lambda$($\lambda$) [W/m$^2$/nm]');
plt.title('Solar Blackbody Spectrum at Earth');

N_y = J_solar_spectra * wavelength/(const.h*const.c); 

# Integrate out d lambda
N = integrate_diff_spectrum(wavelength * 1e9, N_y)
Nph = N/const.c

plt.subplot(1,2,2);
plt.plot(wavelength*1e9, Nph);
plt.xlabel('$\lambda$ [nm]', fontsize=14);
plt.ylabel('Photon Number Density\nN$_{ph}$($\lambda$) [ph/m$^3$]');
plt.title('Solar Blackbody Spectrum at Jupiter');
```

# Inverse Compton Spectrum

## Low-energy Photon Thomson Limit

From *Blumenthal and Gould, 1970* the differential cross section for photon production from inverse Compton scattering (i.e. Compton scattering in the high-energy electron limit) is

$$\frac{dN_{\gamma,e}}{dt d\epsilon} = \frac{\pi r_0^2 c}{2\gamma^4} \frac{N(\epsilon_0)d\epsilon_0}{\epsilon_0^2} \left(2\epsilon ~ln \frac{\epsilon}{4\gamma^2\epsilon_0} + \epsilon + 4\gamma^2\epsilon_0 - \frac{\epsilon^2}{4\gamma^2 \epsilon_0}\right)$$

where the very first term becomes
$$\frac{\pi r_0^2 c}{2\gamma^4} = \frac{3\sigma_T c}{16\gamma^4}$$
by substitution the Thomson cross section in place of the classical electron radius, $r_0$.

$\epsilon_0$ ~ incident photon energy, $[J]$

$\epsilon_1$ ~ scattered photon energy, $[J]$

$N(\nu_0)$ ~ incident photon field number density, $[photons/m^3]$

$\gamma$ ~ electron relativistic factor

$\sigma_T = 8/3~\pi r_0^2 = 6.6524\times10^{-25}$ cm$^2 = 6.6524\times10^{-29}$ m$^2$ ~ Thomson cross section

NB: this spectral integration assumes the Thomson cross section for electrons is valid, that is to say the photon energy after the first Lorentz boost:

$$\gamma ~\epsilon_0 \ll mc^2 = 511 ~keV/c^2$$

and if the above condition is true, the final photon energy in the lab frame gains another factor of $\gamma$ such that the final energy is $\epsilon_1 = \gamma^2 ~\epsilon_0$. Averaged over all solid angles, the average energy after scattering is then

$$\left<\epsilon_1\right> = \frac{4}{3}\gamma^2 ~\epsilon_0$$

## Klein-Nishina High-energy Photon Limit

If we instead use the Klein-Nishina cross section instead of the Thomson cross section, we obtain a result that is true for both Thomson scattering and for higher photon energies:

$$\frac{dN_{\gamma,e}}{dtdE_1} = \frac{3\sigma_T (mc^2) ~c}{4\gamma}\frac{n(\epsilon)d\epsilon}{\epsilon}\left[2q~ln ~q + (1+2q)(1-q) + \frac{1}{2}\frac{(\Gamma_e q)^2}{1+\Gamma_e q}(1-q)\right]$$

where we are writing the scattered photon energy in terms of the initial electron energy

$$\epsilon_1 = \gamma m c^2 E_1$$

and we have the terms

$$\Gamma_e = \frac{4\epsilon\gamma}{mc^2} ~,~~ q = \frac{E_1}{\Gamma_e (1-E_1)}$$

where this expression is valid for all $\Gamma_{e}$ (where $\Gamma_{e} \ll 1 \implies$ Thomson scattering), under the assumption that we have high-energy electrons; $\gamma \gg 1$.

```python
# Thomson cross-section
sigma_T = 6.6524e-29; # m^2

# Electron rest mass 
mc2 = 8.187e-14; # [J] = 511 keV/c^2

# Post-scattering photon energy range and freuency
E  = np.logspace(1, 9, 1000) * const.eV; # eV -> J
nu = E/const.h;

# Initial photon frequency
nu0 = const.c/wavelength;

# Green photon field density at Jupiter
N = 1.5e20 / const.c; # photons/m^2-sec -> photons/m^3

# Electron energy
gamma = 98;


# Thomson limit spectra, Blumenthal and Gould, 1970
I2 = lambda nu0, nu, N, gamma: 3 * sigma_T * const.c / (16 * gamma**4) * N / nu0**2 * \
(2*nu * np.log(nu/(4*gamma**2*nu0)) + nu + 4*gamma**2*nu0 - nu**2/(4*gamma**2*nu0));

# Klein-Nishina limit spectra, Blumenthal and Gould, 1970
def I3(E0, E, N, gamma):
    '''
    E0    : initial photon energy [J]
    E     : photon energy range after scattering [J]
    N     : photon field number density [m^-3]
    gamma : electron energy, E = gamma*mc2
    '''

    E1    = E / ( gamma * mc2 ); 
    
    # Intermediate variable Gamma_e from B&G 1970
    Gamma = 4 * E0 * gamma / mc2;
    
    # Spectral shape parameter
    q = E1 / ( Gamma * (1 - E1) );
    
    Xsec = ( 3 * sigma_T * mc2 * const.c / (4 * gamma) ) * N / E0 * \
    ( 2 * q * np.log(q) + (1 + 2 * q)*(1 - q) + 0.5 * (Gamma * q)**2/(1 + Gamma * q) * (1-q) );
    
    return Xsec;

plt.figure(1,figsize=(12,8)); plt.grid(True, which='both');
plt.figure(2,figsize=(12,8)); plt.grid(True, which='both');

# Loop over initial photon frequency from 250 - 1000 nm
for ind, WL in enumerate(np.logspace(2.4, 3, 10)):
    
    # Conversion to frequency
    initial_photon_freq = const.c / (WL * 1e-9)
    
    # Calculate ICS cross section given monoenergetic electron and photon spectra
    I_tmp = I3(initial_photon_freq*const.h, E, N, gamma);
    
    # Integrate over energy
    #I_tmp = integrate_diff_spectrum(E/const.eV, I_tmp);
    
    # Correct for non-physical cross section ranges
    I_tmp[I_tmp < 0] = 0; 
        
    if ind % 3 == 0:
        label = '$\\nu_0$=%.1f eV' % (const.h*initial_photon_freq/const.eV)
    else:
        label=None;
    plt.figure(1);
    plt.plot(E/const.eV, I_tmp, label=label);
    plt.figure(2);
    plt.plot(E/const.eV, integrate_diff_spectrum(E/const.eV, I_tmp), label=label);

plt.figure(1);
plt.xlabel('Photon Energy after Inverse Compton Scattering [eV]');
plt.ylabel('dI($\\nu$)/dE [photons/s/eV]');
plt.title('Spectra of Monoenergetic Photon Fields from %.0f MeV Electrons' % (gamma*0.511));
plt.legend(); plt.yscale('log'); plt.xscale('log');

plt.figure(2); 
plt.xlabel('Photon Energy after Inverse Compton Scattering [eV]');
plt.ylabel('I($\\nu$) [photons/s]');
plt.title('Spectra of Monoenergetic Photon Fields from %.0f MeV Electrons' % (gamma*0.511));
plt.legend(); plt.yscale('log'); plt.xscale('log'); plt.ylim([1e-1, 1e0]);
```

In order to combine the photon spectrum with an electron spectrum, we can write, again from *Blumenthal and Gould, 1970*:

$$\frac{dN_{Tot}}{dtd\epsilon_1} = \int\int dN_e \frac{dN_{\gamma,e}}{dtd\epsilon_1}$$ 

we'll chose integration over electron gamma factor, i.e. $dN_e = N_e(\gamma) d\gamma$, and use our spectrum from above,

$$\frac{dN_{Tot}}{dtd\epsilon_1} = \int\int N_e(\gamma) d\gamma~ \frac{dN_{\gamma,e}}{dtd\epsilon_1}$$ 

where below we present Jupiter's electron spectra as a function of L-shell and electron $\gamma$.

```python
plt.figure(figsize=(16,12));    

for ind, g in enumerate(np.logspace(0, 2, 10)):

    I_tmp3 = I3(nu0 * const.h, E, Nph, g);
    
    I_tmp3[I_tmp3 < 0] = 0;

    if ind % 3 == 0:
        label = "E$_e$ = %.1f MeV" % (g*0.511);
    else:
        label = None;
    plt.subplot(2,1,1);
    plt.loglog(E/const.eV, I_tmp3, label=label);
    plt.subplot(2,1,2);
    plt.loglog(E/const.eV, integrate_diff_spectrum(E/const.eV, I_tmp3), label=label);

plt.subplot(2,1,1); plt.grid('on', which='both');
plt.ylabel('Spectral Luminosity\n[photons/s/eV]');
plt.title('Solar Spectra Photon Field ICS');
plt.xlabel('Photon Energy after Inverse Compton Scattering [eV]');

plt.legend(); plt.yscale('log'); plt.xscale('log'); plt.xlim([5e0, 1e5]);

plt.subplot(2,1,2); plt.grid('on', which='both');
plt.xlabel('Photon Energy after Inverse Compton Scattering [eV]');
plt.ylabel('$\\frac{dN_{\gamma,e}}{dt}$ [photons/s]');
plt.legend(); plt.yscale('log'); plt.xscale('log'); plt.xlim([5e0, 1e5]);
```

```python
def computeInverseComptonSpectrum(nu0, Nph, E, bandStart, bandStop):
    '''
    nu0: initial photon frequency 
    E  : photon energy array after scattering
    Nph: source photon number density [m^-3] ( spectra Nph(nu0) )
    bandStart, bandStop: energy limits [eV] to compute emitted power between
    
    D&G parameterized with r_grid and gamma_grid
    '''

    # Klein-Nishina limit spectra, Blumenthal and Gould, 1970
    def I3(E0, E, N, gamma):
        '''
        E0    : initial photon energy [J]
        E     : photon energy range after scattering [J]
        N     : photon field number density [m^-3]
        gamma : electron energy, E = gamma*mc2
        '''

        E1    = E / ( gamma * mc2 ); 

        # Intermediate variable Gamma_e from B&G 1970
        Gamma = 4 * E0 * gamma / mc2;

        # Spectral shape parameter
        q = E1 / ( Gamma * (1 - E1) );

        Xsec = ( 3 * sigma_T * mc2 * const.c / (4 * gamma) ) * N / E0 * \
        ( 2 * q * np.log(q) + (1 + 2 * q)*(1 - q) + 0.5 * (Gamma * q)**2/(1 + Gamma * q) * (1-q) );

        return Xsec;

    Iacc   = np.zeros([len(E), len(r_grid)]);
    Pacc   = np.zeros([len(E), len(r_grid)]);
    PaccStartStop = np.zeros([len(E), len(r_grid)]);
    for i in range(0, len(gamma_grid)):

        for r in range(0, len(r_grid)):
            I_tmp = ne_spectra[r,i] * I3(nu0*const.h, E, Nph, gamma_grid[i]);

            I_tmp[I_tmp < 0] = 0;

            I_tmp = integrate_diff_spectrum(E/const.eV, I_tmp);

            Iacc[:,r] += I_tmp;  
            Pacc[:,r] += integrate_diff_spectrum(E/const.eV, I_tmp);

            PaccStartStop[:,r] += integrate_diff_spectrum(E/const.eV, I_tmp);
            PaccStartStop[(E/const.eV < bandStart) | (E/const.eV > bandStop),r] = 0;

    return Iacc, Pacc, PaccStartStop;


def mapICStoLshell(Pacc15, bandStart, bandStop):
    from scipy.ndimage import zoom

    l_grid = np.linspace(0, np.pi/2, len(r_grid));

    flux_grid = np.zeros([len(r_grid)+1, len(r_grid)+1]);
    flux_grid = np.random.poisson(lam=0, size=[len(r_grid)+1, len(r_grid)+1])


    IIacc = np.nansum(Pacc15, axis=0);

    yShift = 2;
    xShift = 1;
    for ind, L in enumerate(np.linspace(1.1, 15.8, len(r_grid))):

        # Dipole coordinates to cartesian
        x_grid = L * np.cos(l_grid)**3;
        y_grid = L * np.cos(l_grid)**2 * np.sin(l_grid);

        flux_grid[(x_grid - len(r_grid)/2).astype(int)-xShift+1, 
                  (y_grid - len(r_grid)/2).astype(int)-yShift]  = IIacc[ind];

        flux_grid[(x_grid - len(r_grid)/2).astype(int)-xShift+1, 
                  (-y_grid - len(r_grid)/2).astype(int)-yShift-1] = IIacc[ind];

        flux_grid[(-x_grid - len(r_grid)/2).astype(int)-xShift, 
                  (y_grid - len(r_grid)/2).astype(int)-yShift]  = IIacc[ind];

        flux_grid[(-x_grid - len(r_grid)/2).astype(int)-xShift, 
                  (-y_grid - len(r_grid)/2).astype(int)-yShift-1] = IIacc[ind];


    Xt,Yt = np.meshgrid(np.linspace(-r_grid[-1], r_grid[-1], len(r_grid)+1),
                        np.linspace(-r_grid[-1], r_grid[-1], len(r_grid)+1));


    plt.figure(figsize=(16,8));
    plt.pcolormesh(Xt-1, Yt-2, 1e7*np.rot90(flux_grid, 1), cmap='jet', norm=matplotlib.colors.LogNorm(),);
                   #vmin=1e8, vmax=5e14);
    cbar=plt.colorbar();
    cbar.set_label(label='Power [ergs/s]');
    #plt.scatter([1], [1], color='r', s=500);
    plt.xlabel('X [R$_J$]'); plt.ylabel('Y [R$_J$]');
    plt.ylim([-11, 11]); 
    plt.xlim([-17,16]);
    plt.title('Total Power emitted between %.1f - %.0f keV' % (bandStart/1000, bandStop/1000));
```
